#### Calculate the sample mean and variance of the variables: price, rooms, surface area and bathrooms (★★★) 

```python
import numpy as np


mean_price = np.mean(ds['price'])
mean_rooms = np.mean(ds['rooms'])
mean_surface = np.mean(ds['surface'])
mean_bathrooms = np.mean(ds['bathrooms'])

var_price = np.var(ds['price'], ddof=1)  
var_rooms = np.var(ds['rooms'], ddof=1)
var_surface = np.var(ds['surface'], ddof=1)
var_bathrooms = np.var(ds['bathrooms'], ddof=1)

print("la media de Price:", mean_price)
print("la Variance de Price:", var_price)
print("la media de Rooms:", mean_rooms)
print("la Variance de Rooms:", var_rooms)
print("la media de Surface:", mean_surface)
print("la Variance de Surface:", var_surface)
print("la media de Bathrooms:", mean_bathrooms)
print("la Variance de Bathrooms:", var_bathrooms)
```

#### Normalize the variable of prices for each population and plot the 4 histograms in the same plot (★★★) 

```python

ds_south = pd.DataFrame(ds_south)
mean_price = ds_south['price'].mean()
std_price = ds_south['price'].std()

ds_south['normalizado'] = ds_south['price'].apply(lambda x: (x- mean_price)/std_price)

#########################

import matplotlib.pyplot as plt


# #"Fuenlabrada","Leganés","Getafe","Alcorcón"

sim_1 = ds_south[ds_south['level5']=="Fuenlabrada"]['normalizado']
sim_2 = ds_south[ds_south['level5']=="Leganés"]['normalizado']
sim_3 = ds_south[ds_south['level5']=="Getafe"]['normalizado']
sim_4 = ds_south[ds_south['level5']=="Alcorcón"]['normalizado']

plt.hist(sim_1, label='Fuenlabrada',alpha=0.6, zorder=3)
plt.hist(sim_2, label='Leganés',alpha=0.5, zorder=4)
plt.hist(sim_3, label='Getafe',alpha=0.8)
plt.hist(sim_4, label='Alcorcón', alpha=0.8 )

plt.legend(loc='upper right')
plt.show()
##########################################################

```

#### Make the same plot for 4 different populations (level5) and rearrange them on the same graph? (★★☆) 
Hint: make a scatter plot of each population using subplots.

```python


# List of populations (level5 values)
populations = ['Getafe', 'Alcorcón', 'Valdemorillo', 'Galapagar']

# Create a new figure
plt.figure(figsize=(10, 6))

# Loop through each population and plot scatter plot
for population in populations:
    df_population = ds[ds['level5'] == population]
    plt.scatter(df_population['surface'], df_population['price'], label=population)

# Add labels and title
plt.xlabel('Superficie')
plt.ylabel('Precio')
plt.title('Relación entre Superficie y Precio para Diferentes Poblaciones')
plt.legend()

# Show the plot
plt.show()






```
